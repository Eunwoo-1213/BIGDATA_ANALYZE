```python
import numpy as np
a = np.arange(1, 13).reshape(3,4)
c = np.arange(1, 6, 2)
a, c
```

### hstack(), column stack()

```python
c[:, None]
```

```python
c[:, np.newaxis]
```

```python
cc = np.expand_dims(c, axis = 1)
cc
```

```python
np.hstack((a,cc)) # = column_stack()
```

```python
np.column_stack((a, c))
```

```python
np.column_stack((a, cc))
```
