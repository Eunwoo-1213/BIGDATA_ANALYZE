```python
# if elif
point = 92
if (90 <= point):
  print('A')
elif (80 <= point):
  print('B')
elif (70 <= point):
  print('C')
elif (60 <= point):
  print('D')
else:
  print('F')
```

```python
# match
value = "apple"
match value:
  case "apple":
    result = "사과"
  case "banana":
    result = "바나나"
print(result)
```

```python
value = "banana"
match value:
  case "apple":
    result = "사과"
  case "banana":
    result = "바나나"

print(result)
```

```python
value = "mango"

match value:
  case "apple":
    result = "사과"
  case "banana":
    result = "바나나"
  case _:
    result = None

print(result)
```

```python
def hi(one):
  print('안녕,', one)

hi('영희')

def hi(one = "친구들"):
  print("안녕,", one)

hi()
```

```python
import math as m
def circle_area(r=1):
  print(m.pi * r**2)

circle_area()
circle_area(4.5)
print(circle_area(r=7.2))
```

```python
# return 구문과 함수 매개변수 지정
import math as m
def circle_area(r=1):
  return m.pi * r**2

print(circle_area(r=7.2))
```

```python
def sum_sub(x, y):
  hap = x + y
  cha = x - y
  return hap, cha

sum_sub(10, 3)
```
