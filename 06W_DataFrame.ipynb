### DataFrame 다양한 참조

```python
pf = pd.DataFrame(
    data=[[25, 35, 8, 18],
          [18, 27, 10, 20],
          [17, 17, 10, 19],
          [12, 22, 9, 20],
          [22, 34, 8, 16]],
    index=['윤일형', '강수희', '홍소희', '유한빈', '신수빈'],
    columns=['중간', '기말', '과제', '출석']
)
pf
```

```python
# 열 참조
print(pf['중간'])
print()

print(pf.중간)
print()

print(pf['출석'])
print()

print(pf[['출석']])
print()

print(pf[['기말', '출석', '중간']])
print()

print(pf[['기말']])
```

```python
#행 참조
import pandas as pd

pf = pd.DataFrame(
    data=[[25, 35, 8, 18],
          [18, 27, 10, 20],
          [17, 17, 10, 19],
          [12, 22, 9, 20],
          [22, 34, 8, 16]],
    index=['윤일형', '강수희', '홍소희', '유한빈', '신수빈'],
    columns=['중간', '기말', '과제', '출석']
)
pf
```

```python
print(pf['중간'] > 20)
print()

print(pf[pf['중간'] > 20])
print()

print(pf[pf['출석'] >= 19])
print()

print(pf[(pf['출석'] >= 19) & (pf['기말'] >= 20)])
```
